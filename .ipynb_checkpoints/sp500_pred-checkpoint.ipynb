# Predicting price of S&P 500 five days in advance

Here's my database which I have collected from yfinance API. It's S&P 500 price info from the 1950s and technical indicators adx, rsi and stoch calculated from the price.

```python
import sqlite3
import pandas as pd

database = sqlite3.connect("sp500_megadb.sqlite")
df = pd.read_sql_query("SELECT * FROM prices", database)

df["name"] = "sp500"

df
```

I'm going to make lagged features of all the features so the model is actually using 10 datapoints from the past to predict the price five days into the future.

```python
def make_lags(name, feature, lags):
    return pd.concat(
    {
        f'{name}_lag_{i}': feature.shift(i)
        for i in range(1, lags + 1)
    },
    axis = 1)

df = df.join(make_lags("adx", df.adx, lags=10))
df = df.join(make_lags("dmp", df.dmp, lags=10))
df = df.join(make_lags("dmn", df.dmn, lags=10))
df = df.join(make_lags("rsi", df.rsi, lags=10))
df = df.join(make_lags("slowd", df.slowd, lags=10))
df = df.join(make_lags("slowk", df.slowk, lags=10))
df = df.join(make_lags("price", df.close, lags=10))
df = df.join(make_lags("high", df.high, lags=10))
df = df.join(make_lags("low", df.low, lags=10))

df["target"] = df["close"].shift(-5)

df.dropna(inplace=True)

X = df.drop(["name", "target", "date", "id", "close", "low", "high"], axis=1)
y = df["target"]

X
```

I'm splitting the data into train and test sets. Test size 5% is going to be enough because even 5% is gonna be ~3 years of data.

Next I'm gonna scale all the features to have a 0 as the average so the model is gonna have an easier time.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05, shuffle=False, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

Then to the model. I'm using Adam as the optimizer. The model is gonna have two hidden layers, that should be enough. Numbers of nodes is approx. 1/3 of the input_dimension. The activation function is relu since we are trying to predict a floating point number and for the same reason loss funtion is MSE.

```python
import keras
from tensorflow.keras.optimizers import Adam

opt = Adam(learning_rate=0.001)

model = keras.models.Sequential([
    keras.layers.LSTM(50, return_sequences=True, input_shape=(X_train_scaled.shape[1], 1)),
    #Dropout(0.20LSTM
    keras.layers.LSTM(50, return_sequences=True),
    #Dropout(0.20LSTM
    keras.layers.LSTM(50),
#    keras.layers.LSTM(50, return_sequences=True, input_shape=(X_train_scaled.shape[1], 1)),
#    Dropout(0.2),
#    keras.layers.LSTM(50, return_sequences=True),
#    Dropout(0.2),
#    keras.layers.LSTM(50),
#    Dropout(0.2),
    keras.layers.Dense(1)
])

model.compile(optimizer=opt, loss='mean_squared_error')
res = model.fit(X_train_scaled, y_train, epochs=300, shuffle=False)
```

```python
y_pred = model.predict(X_test_scaled)
```

Let's plot the loss.

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

plt.figure(figsize=(14,6))
sns.lineplot(data=(res.history["loss"]))
plt.xlabel("epoch")
plt.ylabel("loss")
```

Then I'll plot the model's predictions compared to the true values.

```python
df_plot = pd.DataFrame(y_pred, columns=["y_pred"])
df_plot['y_true'] = y_test.to_numpy()

plt.figure(figsize=(20,6))
fig = sns.lineplot(data=df_plot)
fig.lines[1].set_linestyle("-")
plt.show()
```
